```python
import pandas as pd
import numpy as np

# --- 1. Define the Loading Function ---
def load_and_explore_data(filepath):
    """Loads the CSV, displays basic info, and the first few rows."""
    
    # 1. Loading
    try:
        # Tries to read the file using the path passed to the function
        df = pd.read_csv(filepath, low_memory=False) 
    except FileNotFoundError:
        # Error handling if the path is wrong
        print(f"Error: File not found at {filepath}. Please check the DATA_PATH.")
        return None
    
    print("--- üìù DataFrame General Information (Check for NaNs & Types) ---")
    df.info()
    
    print("\n--- üßê First Rows ---")
    print(df.head())
    
    print("\n--- ‚úÖ DATASET SIZE ---")
    print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
    
    return df

# --- 2. Execution ---
# NOTE: We assume the file is in 'data/raw/test.csv' relative to your notebook.
DATA_PATH = 'data/raw/test.csv' 
df_aero_vista = load_and_explore_data(DATA_PATH)
```
